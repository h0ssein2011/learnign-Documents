```python
import pandas as pd
import numpy as np
import collections
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/mattharrison/datasets/master/data/2020-jetbrains-python-survey.csv'
jb = pd.read_csv(url)
jb.head()
```

```python
def tweak_jb(df):
    return (df
    .rename(columns={v:v.replace('.','_') for v in df.columns})
    .assign(age= lambda df_: df_.age.str.slice(0,2).astype(float))
    .assign(team_size= lambda df_: df_.team_size.replace({np.nan:0, '2-7 people':5, '8-12 people':10, '21-40 people':30, '13-20 people':15,'More than 40 people':50}))
    .assign(python3_ver=lambda df_:df_.python3_version_most
                     .str.replace('_', '.').str.extract(r'(\d\.\d)')
                     .astype(float),
                use_python_most=lambda df_:df_.use_python_most
                     .fillna('Unknown'))
    )
    
jb2 = tweak_jb(jb)
jb2.head()  
```

```python
pd.crosstab(index = jb2.country_live,columns = jb2.age)
```

```python
# pd.crosstab(index = jb2.country_live,columns = jb2.age,margins=True)
pd.crosstab(index = jb2.country_live,columns = jb2.age,normalize='columns',margins=True)
```

```python
(pd.crosstab(index=[jb2.country_live,jb2.age],
columns = [jb2.use_python_most ,jb2.python3_ver])
.loc[['United States'],['Data analysis','Web development']]
.style.background_gradient(cmap='viridis',axis=None)
)
```
