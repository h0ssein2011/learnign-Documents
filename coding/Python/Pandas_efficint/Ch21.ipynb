```python
import pandas as pd
import numpy as np
import collections
```

```python
url = 'https://raw.githubusercontent.com/mattharrison/datasets/master/data/2020-jetbrains-python-survey.csv'
df = pd.read_csv(url)
df.head()
```

```python
for c in sorted(df.columns):
    print(c,'-->', df[c].dtype)
```

```python
countr = collections.defaultdict(list)
for col in sorted(df.columns):
    period_count = col.count('.')
    if period_count >= 2 :
        part_end = 2
    else:
        part_end = 1
    parts = col.split('.')[:part_end]
    countr['.'.join(parts)].append(col)
# print(countr)
```

```python
uniq_cols = []
for k,v in countr.items():
    if len(v) == 1:
        uniq_cols.append(v[0])
        
```

```python
(df
[uniq_cols]
.rename(columns = lambda c:c.replace('.','_'))
.age
.value_counts(dropna=False)
)
```
