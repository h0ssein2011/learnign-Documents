```python
import pandas as pd
import numpy as np
import collections
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/mattharrison/datasets/master/data/2020-jetbrains-python-survey.csv'
jb = pd.read_csv(url)
jb.head()
```

```python
def tweak_jb(df):
    return (df
    .rename(columns={v:v.replace('.','_') for v in df.columns})
    .assign(age= lambda df_: df_.age.str.slice(0,2).astype(float))
    .assign(team_size= lambda df_: df_.team_size.replace({np.nan:0, '2-7 people':5, '8-12 people':10, '21-40 people':30, '13-20 people':15,'More than 40 people':50}))
    )
    
jb2 = tweak_jb(jb)
jb2.head()  
```

```python
(jb2.
pivot_table(index='country_live',columns='employment_status',values='age',aggfunc='mean')
)
```

```python
#use crostab instead of pivot_table
(pd.
crosstab(index=jb2.country_live,columns=jb2.employment_status,values=jb2.age,aggfunc='mean')
)
```

```python
(jb2.
groupby(['country_live','employment_status'])
.age
.mean()
.unstack()
)
```

```python
# use custom function to get the percentage of each countrry use EMACs
# jb2.ide_main.value_counts(normalize=True) * 100

def per_Emacs(ser):
    return ser.str.contains('Emacs').sum() * 100  / len(ser)
```

```python
jb2.pivot_table(index='country_live',values='ide_main',aggfunc=per_Emacs)
```

```python
(jb2
.groupby('country_live')
[['ide_main']]
.agg(per_Emacs)
)
```

```python
#remove highracy columns
(jb2
.groupby('country_live')
.agg(age_min=('age','min'),age_max=('age','max'),team_size_mean=('team_size','mean'))
.reset_index()
)#.plot(x='country_live',y=['age_max'],kind='bar',figsize=(25,5))
```

```python
# return max and min age of each country and editor they used
(jb2
.pivot_table(index=['country_live','ide_main'],values='age',aggfunc=['min','max'])

)
```

```python
# return the average team size of each country and editor they used with group by

"""
when grouping with multiple categorical columns with high cardinality. You can generate a very large (and sparse) result!
we can use the observed option.

"""
# (jb2
# .assign(country_live2=lambda df_: df_.country_live.astype('category'))
# .groupby(['country_live2','ide_main'])
# ['age']
# .agg(['min','max','mean'])
# )


(jb2
.assign(country_live2=lambda df_: df_.country_live.astype('category'))
.groupby(['country_live2','ide_main'],observed=True)
['age']
.agg(['min','max','mean']))

```

```python
url = 'https://raw.githubusercontent.com/h0ssein2011/datasets/main/vehicles.csv'
cols = ['id','city08', 'comb08', 'highway08', 'cylinders','displ', 'drive', 'eng_dscr', 'fuelCost08','make', 'model', 'trany', 'range', 'createdOn','year']
vehicles = pd.read_csv(url,low_memory=False,usecols=cols)
vehicles = vehicles[cols]
vehicles.head()
```

```python
# 1. Group by a categorical column and take the mean of the numeric columns.
(vehicles
# .groupby('drive')['city08']
# .mean()
.pivot_table(index='drive',values='city08',aggfunc='mean')
)
```

```python
# Group by a categorical column and take the mean and max of the numeric column
(vehicles
# .groupby('drive')['city08']
# .agg(['mean','max'])
.pivot_table(index='drive',values='city08',aggfunc=['mean','max'])
)
```

```python
# Group by a categorical column and apply a custom aggregation function that calculate mode of the numeric columns.
def mode(ser):
    return ser.mode()[0]
(vehicles
# .groupby('drive')['cylinders']
# .agg(mode)
.pivot_table(index='drive',values='cylinders',aggfunc=mode)
)
```

```python
# Group by two categorical columns and take the mean of the numeric columns.
(vehicles
# .groupby(['drive','cylinders'])['city08']
# .mean()
.pivot_table(index=['drive','cylinders'],values='city08',aggfunc='mean')
)
```

```python
## Group by binned numeric column and take the mean of the numeric columns.
def bin_mean(ser):
    quantile = ser.quantile(0.25)
    return ser[ser < quantile].mean()

(vehicles
.groupby('city08')['city08']
.agg(bin_mean)
# .mean()

)
```
