```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python


# Load the time series data
df = pd.read_csv('airline-passengers.csv', index_col='Month', parse_dates=True)

# Define the model
model = sm.tsa.Holt(df['Passengers'], exponential=False, initialization_method="estimated")

# Fit the model
fit = model.fit()

# Generate a forecast for the next 12 periods
forecast = fit.forecast(12)

# Plot the original data and the forecast
plt.plot(df.index, df['Passengers'], label='Observed')
plt.plot(forecast.index, forecast, label='Forecast')
plt.legend()
plt.show()
```
