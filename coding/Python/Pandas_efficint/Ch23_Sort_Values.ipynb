```python
import pandas as pd
import numpy as np
```

```python
url = 'https://github.com/mattharrison/datasets/raw/master/data/siena2018-pres.csv'
pres = pd.read_csv(url,index_col=0)
pres.head()
```

```python
(pres
.sort_values(by='Party')
)
```

```python
(pres
.sort_values(by=['Party','Bg'],ascending=[False,True])
)
```

```python
(pres
.President
.str.split()
.apply(lambda x: x[-1])
)
```

```python
#sort by last name
(pres
.sort_values(by='President',key=lambda name_ser: name_ser
.str.split()
.apply(lambda x: x[-1]))
)
```
