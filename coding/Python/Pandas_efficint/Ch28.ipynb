```python
import pandas as pd
import numpy as np
import collections
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/mattharrison/datasets/master/data/2020-jetbrains-python-survey.csv'
jb = pd.read_csv(url)
jb.head()
```

```python
def tweak_jb(df):
    return (df
    .rename(columns={v:v.replace('.','_') for v in df.columns})
    .assign(age= lambda df_: df_.age.str.slice(0,2).astype(float))
    .assign(team_size= lambda df_: df_.team_size.replace({np.nan:0, '2-7 people':5, '8-12 people':10, '21-40 people':30, '13-20 people':15,'More than 40 people':50}))
    )
    
jb2 = tweak_jb(jb)
jb2.head()  
```

```python
(jb2
.groupby('country_live')
.age
.transform('size')
)
```

```python
# Add a column for size of respondants in each country (number of people)
(jb2
.assign(country_resposes = (jb2
.groupby('country_live')
.age
.transform('size')
))
)
```

```python
# filter countries less than median
countries_to_remove = (jb2
.country_live
.value_counts()
.lt(126.5)
.index
)

```

```python
(jb2
.query('~country_live.isin(@countries_to_remove)')

)
```

```python
# 1. Add a new column that is the sum of a numeric column that was grouped by a string column.
# 2. Filter out the rows that have less than 3 entries when grouped by a string column.
url = 'https://raw.githubusercontent.com/h0ssein2011/datasets/main/vehicles.csv'
cols = ['id','city08', 'comb08', 'highway08', 'cylinders','displ', 'drive', 'eng_dscr', 'fuelCost08','make', 'model', 'trany', 'range', 'createdOn','year']
vehicles = pd.read_csv(url,low_memory=False,usecols=cols)
vehicles = vehicles[cols]
print(vehicles.shape)
vehicles.head()



```

```python
high_usag_makers = (vehicles
.groupby('make')
.highway08
.transform('sum')
.gt(100)
*1
)

(vehicles
.assign(is_high_usag_makers=high_usag_makers)
# .is_high_usag_makers
# .value_counts()

)
```
