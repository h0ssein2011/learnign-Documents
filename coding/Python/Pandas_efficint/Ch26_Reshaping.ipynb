```python
import pandas as pd
import numpy as np
import collections
```

```python
url = 'https://raw.githubusercontent.com/mattharrison/datasets/master/data/2020-jetbrains-python-survey.csv'
jb = pd.read_csv(url)
jb.head()
```

```python
# #replace the NaN with 0 
(jb.
filter(like='job.role.')
.where(jb.isna(),1)
.fillna(0)

)
```

```python
#Collapse the job roles in to a single column 
# #replace the NaN with 0 
job  = (jb.
filter(like='job.role.')
.where(jb.isna(),1)
.fillna(0)
.idxmax(axis=1)
#replace the job.role. with ''
.str.replace('job.role.','',regex=False)
)
```

```python
dum = pd.get_dummies(job)
dum.head()
```

```python
#undoing dummy columns
dum.idxmax(axis=1)
```

```python
i , j = np.where(dum)
pd.Series(dum.columns[j],index=i)
```
