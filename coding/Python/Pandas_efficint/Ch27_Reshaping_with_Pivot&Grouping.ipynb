```python
import pandas as pd
import numpy as np
import collections
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/mattharrison/datasets/master/data/2020-jetbrains-python-survey.csv'
jb = pd.read_csv(url)
jb.head()
```

```python
def tweak_jb(df):
    return (df
    .rename(columns={v:v.replace('.','_') for v in df.columns})
    .assign(age= lambda df_: df_.age.str.slice(0,2).astype(float))
    .assign(team_size= lambda df_: df_.team_size.replace({np.nan:0, '2-7 people':5, '8-12 people':10, '21-40 people':30, '13-20 people':15,'More than 40 people':50}))
    )
    
jb2 = tweak_jb(jb)
jb2.head()  
```

```python
(jb2.
pivot_table(index='country_live',columns='employment_status',values='age',aggfunc='mean')
)
```

```python
#use crostab instead of pivot_table
(pd.
crosstab(index=jb2.country_live,columns=jb2.employment_status,values=jb2.age,aggfunc='mean')
)
```

```python
(jb2.
groupby(['country_live','employment_status'])
.age
.mean()
.unstack()
)
```

```python
# use custom function to get the percentage of each countrry use EMACs
# jb2.ide_main.value_counts(normalize=True) * 100

def per_Emacs(ser):
    return ser.str.contains('Emacs').sum() * 100  / len(ser)
```

```python
jb2.pivot_table(index='country_live',values='ide_main',aggfunc=per_Emacs)
```

```python
(jb2
.groupby('country_live')
[['ide_main']]
.agg(per_Emacs)
)
```
