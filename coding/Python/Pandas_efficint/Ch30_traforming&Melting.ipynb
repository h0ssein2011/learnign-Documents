# Stacking and unstacking data
# stack data


```python
url = 'https://raw.githubusercontent.com/mattharrison/datasets/master/data/2020-jetbrains-python-survey.csv'
jb = pd.read_csv(url)
jb.head()
```

```python
def tweak_jb(df):
    return (df
    .rename(columns={v:v.replace('.','_') for v in df.columns})
    .assign(age= lambda df_: df_.age.str.slice(0,2).astype(float))
    .assign(team_size= lambda df_: df_.team_size.replace({np.nan:0, '2-7 people':5, '8-12 people':10, '21-40 people':30, '13-20 people':15,'More than 40 people':50}))
    .assign(python3_ver=lambda df_:df_.python3_version_most
                     .str.replace('_', '.').str.extract(r'(\d\.\d)')
                     .astype(float),
                use_python_most=lambda df_:df_.use_python_most
                     .fillna('Unknown'))
    )
    
jb2 = tweak_jb(jb)
jb2.head()  
```

```python
#create multilevel index
(jb2 
.groupby(['country_live','are_you_datascientist'])
.size()
)
```

```python
#unstack the data
(jb2 
.groupby(['country_live','are_you_datascientist'])
.size()
.unstack()
)
```

```python
#craete multilevel index
(jb2 
.groupby(['country_live','are_you_datascientist'])
.size()
# .unstack(0)
.unstack('country_live')
)
```

```python
(jb2

.pivot_table(index='country_live',aggfunc={'age':['min','max'],'team_size':['min','max']})
# .stack(0)
.unstack()
# .swaplevel()
)
```

```python
(jb2
.groupby(['country_live','age'])
.mean()
.reset_index()
)
```
